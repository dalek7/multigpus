```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import tensorflow as tf
import keras
```

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
Y = iris.target
print(X.shape)
print(Y.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
Y_1hot = enc.fit_transform(Y.reshape(-1, 1)).toarray()
print(Y[0], " -- one hot enocding --> ", Y_1hot[0])
print(Y[50], " -- one hot enocding --> ", Y_1hot[50])
print(Y[100], " -- one hot enocding --> ", Y_1hot[100])
```

```python
from keras.backend import tensorflow_backend as K
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
K.set_session(tf.Session(config=config))

from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(4, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))

print(keras.__version__)
from keras import backend as K
K.tensorflow_backend._get_available_gpus()
```

```python
from keras.utils.training_utils import multi_gpu_model
model = multi_gpu_model(model, gpus=2)
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y_1hot, epochs=1000, batch_size=1024)
```

```python
Y_pred = model.predict(X)
```

```python
import numpy as np
Y_pred2 = np.argmax(Y_pred, axis=1)
for i in range(np.shape(Y_pred)[0]):
    print('{} ---> {} '.format(Y_pred[i], Y_pred2[i]))
```

```python
print('Misclassified Samples:', (Y != Y_pred2).sum())
```
